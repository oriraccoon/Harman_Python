```python
# 비교 연산자 : 두 개의 값을 비교
# 비교 연산의 결과는 항상 논리 타입 (boolean type)
a = 3
b = 7
print(a > b) # False
print(a >= b) # False
print(a < b) # True
print(a <= b) # True
print(a == b) # False
print(a != b) # True
```

```python
# 논리 연산자
# not, and, or
# not 논리 -> True <-> False
# 논리 and 논리
# 논리 그리고 논리
# 오늘은 2월 5일입니다 : True
# 오늘은 흐리다 
# 오늘은 2월 5일이다 그리고 오늘은 흐리다
# -> True 
# 오늘은 2월 5일이다 그리고 오늘은 흐리지 않다
# -> False 
# 오늘은 2월 6일이다 그리고 오늘은 흐리다
# -> False
# 오늘은 2월 6일이다 그리고 오늘은 흐리지 않다
# -> False
# 논리 곱
# True -> 1
# False -> 0
# 1 x 1 = 1
# 1 x 0 = 0
# 0 x 1 = 0
# 0 x 0 = 0
# True and True -> True 
# True and False -> False
# False and True -> False
# False and False -> False 
c = 10
d = 7
print(not(c > d)) # False
print(5 > 3 and 10 > 7) # True
print(5 < 3 and 10 > 7) # False

```

```python
# or 연산자
# 논리합
# True or True -> True
# True or False -> True
# False or True -> True
# False or False -> False
# 1 + 1 -> 2 (True)
# 1 + 0 -> 1 (True)
# 0 + 1 -> 1 (True)
# 0 + 0 -> 0 (False)

# A + 1 -> 1
```

```python
# 삼항 연산자 (조건 연산자)
# a if 조건식 else b
# 조건식 : 결과가 논리!! 
# 조건식의 결과가 True(참)이면 a(if절) 실행
# 조건식의 결과가 False(거짓)이면 b(else절) 실행
score = int(input("점수를 입력하세요 :"))
result = "합격" if score >= 60 else "불합격"
print(result)
```

```python
# 1
number1 = int(input("첫 번째 정수 입력>>"))
number2 = int(input("두 번째 정수 입력>>"))
sub = number1 - number2 if number1 > number2 else number2 - number1
print(f"두 수의 차 : {sub}")

# 2
number1 = int(input("첫 번째 정수 입력>>"))
number2 = int(input("두 번째 정수 입력>>"))
result = number1 - number2
sub = result if result > 0 else -result
print(f"두 수의 차 : {sub}")
```
