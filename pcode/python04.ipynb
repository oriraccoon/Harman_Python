```python
# 사용자로부터 정수를 입력 받아
# 누적 합하여 출력하는 코드를 완성하시오
# 단, 사용자가 -1을 입력하면 프로그램을 종료하시오
```

```python
total = 0 # 누적합을 저장하는 변수!!

# 대개 반복횟수가 정해져 있지 않을 때 
# while문을 사용하자
# while True -> 무한 반복
while True:
    num1 = int(input("정수를 입력하세요 :"))
    if num1 == -1:
        print("프로그램 종료")
        break # 반복문 종료!! (단, 가장 가까운)
    else:
        total += num1
        print(f"누적 합 :{total}")
```

```python
total = 0 # 누적합을 저장하는 변수!!

num1 = int(input("정수를 입력하세요 :"))

while num1 != -1:
    total += num1
    print(f"누적 합 :{total}")
    
    num1 = int(input("정수를 입력하세요 :"))

print("프로그램 종료")
```

```python
# while문을 사용하여
# "파이썬 최고!!"를 13번 출력

count = 0 # 출력 횟수를 나타내는 변수
while count != 13:
    print("파이썬 최고!!")
    count += 1
```

```python
curr_weight = int(input("현재 몸무게 :"))
goal_weight = int(input("목표 몸무게 :"))
week = 1
while curr_weight > goal_weight:
    # 사용자로부터 뺄 몸무게 입력!!
    loss_weight = int(input(f"{week}주차 감량 몸무게 :"))
    curr_weight -= loss_weight
    week += 1
print(f"{curr_weight}kg 달성!! 축하합니다!")
```

```python
week = 1
curr_weight = int(input("현재 몸무게 :"))
goal_weight = int(input("목표 몸무게 :"))

while True:
    loss_weight = int(input(f"{week}주차 감량 몸무게 :"))
    curr_weight -= loss_weight
    week += 1
    if curr_weight <= goal_weight:
        print(f"{curr_weight}kg 달성!! 축하합니다")
        break
```

```python
# 1 ~ 50까지 랜덤하게 숫자를 추출
# 랜덤 라이브러리!!
import random

# randint(start, end) : start <= 랜덤수 <= end
num = random.randint(1, 50)
# print(num)

x = int(input("숫자를 입력하세요 >>"))

while x != num:
    if x > num:
        print(f"{x}보다 작은 수 입니다.")
    else:
        print(f"{x}보다 큰 수 입니다.")
        
    x = int(input("숫자를 입력하세요 >>"))

print("정답을 맞추셨습니다")

    
```

```python
# 1 ~ 100까지 출력! 
# 이 중 3의 배수만 출력!!
# 1 ~ 100까지 중 3의 배수의 누적합!! 출력!
total = 0 # 누적합
for i in range(1, 101):
    if i % 3 == 0:
        total += i
        # print(i, end=" ")
print(total)
```

```python
list1 = [4, 5, 2, 1, 99, 15, 2, 7, 27]
# 가장 작은 값 출력!!
min_value = list1[0]

# 모든 요소와 min_value 비교
# 만약, 요소가 min_value보다 작다면 교체!!

for value in list1:
    if value < min_value:
        min_value = value
print(min_value)
```

```python
# 기사 시험의 합격 점수는 60점 이상입니다
# 다음의 5명의 사람의 점수가 주어졌을때
# 다음과 같이 출력하시오
tuple1 = (90, 45, 70, 60, 55)
number = 1
for score in tuple1:
    result = "불합격"
    if score >= 60:
        result = "합격"
    print(f"{number}번 학생은 {result}입니다")

    number += 1
```

```python
tuple1 = (90, 45, 70, 60, 55)

# range(5) -> 0, 1, 2, 3, 4
for i in range(len(tuple1)):
    if tuple1[i] >= 60:
        result = "합격"
    else:
        result = "불합격"
    print(f"{i + 1}번 학생은 {result}입니다")
```
