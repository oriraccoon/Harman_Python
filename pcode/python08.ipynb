```python
menu = ["파스타", "햄버거", "제육볶음"]
for i in range(len(menu)):
    print(f"{i + 1}번째 메뉴 : {menu[i]}")

# for m in menu -> 요소를 직접 꺼내기
# for i in range -> 인덱스를 통해 접근

# enumerate() -> 인덱스와 요소 동시에 반환
for i, m in enumerate(menu):
    print(f"{i + 1}번째 메뉴 : {m}")
    
# 1번째 메뉴 : 파스타
# 2번째 메뉴 : 햄버거
# 3번째 메뉴 : 제육볶음
```

### 예외 처리!!

```python
# 코드 작성 시 발생할 수 있는 예외 상황들에 대해
# 프로그램의 안정성과 신뢰성을 높이기 위해 사용
# try : 오류가 발생할 수 있는 코드
# except : 오류 발생 시 처리할 코드
# finally : 오류 발생 여부와 관계 없이 항상 실행
try:
    number = int(input("정수를 입력하세요 :"))
    print(10 / number)
# 그냥 except 사용 시
# 모든 Error 클래스에 대해 탐색!!
except ZeroDivisionError:
    print("0으로 나눌 수 없습니다")
except ValueError:
    print("정수를 입력해주세요")
finally:
    print("프로그램 종료")
```

### shutil
- shell utilites
- 파일 복사, 이동, 삭제 등 고급 기능 제공
- os보다 강력한 기능

```python
import shutil
import os
# shutil.copy() -> 파일 복사
# 복사된 파일은 내용과 권한이 동일
# 권한 : 나, 그룹, 다른 사용자
# 읽고, 쓰고, 실행에 관한 내용

shutil.copy("sample.txt", "backup.txt")
#os.listdir()
```

```python
# shutil.copytree() 폴더 복사!!
shutil.copytree("practice_folder", "backup_folder")
#print(os.listdir())
```

```python
# 파일 이동
# shutil.move("이전 경로", "새로운 경로")
# 이전 파일은 삭제!!

# sample2.txt를 backup_folder로 옮겨보자
shutil.move("sample2.txt", "backup_folder/sample2.txt")
```

```python
os.listdir("backup_folder")
```

```python
# practice_folder를 backup_folder안으로 이동하자
shutil.move("practice_folder", "backup_folder")
print(os.listdir("backup_folder"))
```

```python
# 파일 삭제 
# os.remove()
os.remove("backup_folder/renamed_test.txt")
os.listdir("backup_folder")
```

```python
# 폴더 강제 삭제
# shutil.rmtree()
# 모든 내용이 삭제되므로
# 윈도우키 + . => 이모지
# ⚠️⚠️반드시 주의해서 사용⚠️⚠️
shutil.rmtree("backup_folder")
os.listdir()
```

```python
# 3. exam.txt파일을 exam 폴더로 이동!!
shutil.move("exam.txt", "exam/exam.txt")
os.listdir("exam")
```

```python
# 5. test폴더 내 모든 파일을 삭제!!
list_test = os.listdir("test")
print(list_test)
for f in list_test:
    path = os.path.join("test", f)
    os.remove(path)
```

```python
# 6. test폴더를 삭제!!
os.rmdir("test")
os.listdir()
```

```python
# 1. sample.txt파일을 exam.txt로 복사!!
shutil.copy("sample.txt", "exam.txt")

# 2. exam 폴더 생성!!
if not os.path.exists("exam"):
    os.mkdir("exam")
else:
    print("폴더가 존재합니다")

# 3. exam.txt파일을 exam 폴더로 이동!!
shutil.move("exam.txt", "exam/exam.txt")

# 4. exam폴더 이름을 test로 변경!!
os.rename("exam", "test")

# 5. test폴더 내 모든 파일을 삭제!!
list_test = os.listdir("test")
print(list_test)
for f in list_test:
    path = os.path.join("test", f)
    os.remove(path)

# 6. test폴더를 삭제!!
os.rmdir("test")
```
