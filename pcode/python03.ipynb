```python
##### # 90점 이상이면 -> "A"
# 80 ~ 89점 -> "B"
# 70 ~ 79점 -> "C"
# 60 ~ 69점 -> "D"
# ~ 59점 -> "F"
score = 75
if score >= 90:
    print("A")
elif score >= 80:
    print("B")
elif score >= 70:
    print("C")
elif score >= 60:
    print("D")
else:
    print("F")
```

```python
list1 = [2, 5, 7, 9, 10]
print(list1)
# list1의 첫 번째 요소인 2를 가리키고 싶다
# 리스트명[해당 인덱스]
# python 리스트는 0부터 인덱스 부여
print(list1[0])
# list1의 2번째 값과 마지막 값을 더해서 출력!!
print(list1[1] + list1[-1])
```

```python
list2 = [1, 2, 3, ["a", "b", "c"]]
# a를 출력!!
print(list2[3][0])
```

```python
# list slicing
# 리스트명[시작 인덱스 : 끝 인덱스]
# 단, 끝 인덱스는 포함 X
list3 = [0, 1, 2, 3, 4]
# [0, 1, 2] 출력!!
print(list3[0:3])
print(list3[:4])
```

```python
# 리스트 더하기 가능
list4 = [1, 2, 3]
list5 = [4, 5, 6, 7]
print(list4 + list5)
```

```python
list8 = [17, 23, 27, 14, 36, 13]
print(list8[2:4])
# data type을 확인하기 위한 함수
# type(값)
# 해당 값의 타입을 반환
print(type(list8[2:4]))
print(type(list8[-1]))
```

```python
## 리스트 값 삭제 1
# del 리스트명[인덱스]
# del 리스트명[시작 인덱스 : 끝 인덱스 + 1]
list9 = [1, 2, 3, 4, 5]
print(list9)

del list9[2]
print(list9)

del list9[2:]
print(list9)
```

```python
print(list8)

# 정렬 (Sorting)
# 리스트명.sort() <- 오름차순(ascending)
list8.sort()
print(list8)

# 내림차순 (descending)
# 오름차순 정렬 후 reverse()를 통해 뒤집기
list8.reverse()
print(list8)
```

```python
# 리스트명.index(값) -> 해당 값의 인덱스 반환
list11 = ["a", "b", "c", "d", "e", "f"]
print(list11.index("c"))

# 리스트명.pop() -> 마지막 요소 반환 후 제거
print(list11.pop())
print(list11)
```
