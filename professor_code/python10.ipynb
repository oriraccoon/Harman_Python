```python
# loc 슬라이싱
# loc 슬라이싱은 [시작:끝] -> 시작 포함, 끝 포함
df.loc[3:5]

df.loc[3:5, "Name":"Age"] # 시작 포함, 끝 포함

# 모든 행, 컬럼명 : "Name"
df.loc[:, "Name"] # -> df["Name"]과 동일
```

```python
# Boolean Indexing
# 비교 연산자를 통해 
# True, False를 가지는 Series를 얻을 수 있다
# loc만 가능!!
df.loc[df["Age"] >= 30]

# "Age"가 30이상이고, "City"가 "Busan"인 사람을 구하시오
# Boolean Indexing에서 and와 or는 '&'와 '|'를 사용한다
df.loc[(df["Age"] >= 30) & ((df["City"] == "Busan") | (df["City"] =="Seoul"))]
```

```python
# iloc (Integer Location)
# loc와 기능이 거의 흡사
# 행과 열을 모두 정수를 사용해야 한다
df.iloc[3]
print(df.iloc[3, 1])
print(df.iloc[3:6]) # 시작 포함, 끝 포함 X
print(df.iloc[3:6, 0:2])

# df.loc[:, "Name"] -> df["Name"]
# df.iloc[:, 2] -> df[2] (불가능)
print(df.iloc[:, 2])

# iloc는 boolean indexing 불가능!!
```

```python
# 그룹화 예제 데이터
data = {
    "Department": ["HR", "IT", "IT", "Finance", "HR", "Finance"],
    "Name": ["Alice", "Bob", "Charlie", "David", "Eve", "Frank"],
    "Salary": [5000, 6000, 6500, 7000, 5200, 7100]
}

df3 = pd.DataFrame(data)
df3
```

```python
# 부서별 평균 급여 계산
# 1. 부서를 그룹화
# 2. 급여를 인덱싱
# 3. 평균 함수 적용
df3.groupby("Department")["Salary"].mean()
```

```python
# 부서별 인원 수
# 1. 부서를 그룹화
# 2. 인덱싱은 마음에 드는 컬럼!!
#    일반적으로 이름, 인덱스
# 3. count()

df3.groupby("Department")["Name"].count()
```

```python
# 그룹화 시킨 후 사용하는 함수
# 그룹(group) 함수, 집계(aggregate) 함수
```

```python
# 부서별 인원수, 급여 합계
# 1. 부서를 그룹화
# 2. 급여 인덱싱
# 3. count(), sum()
# 동시에 2개 이상 그룹함수 들어올 경우
# agg를 통해 엮어주자

df3.groupby("Department")["Salary"].agg(["count", "sum"])
```

```python
# 부서별 급여의 최댓값과 최솟값을 구해주세요
# 최댓값 : max(), 최솟값 : min()
df3.groupby("Department")["Salary"].agg(["max", "min", "mean", "sum"])
```

```python
# 급여가 6000이상인 사람만 필터링!!
# 부서별 평균 급여 계산
# 1. Boolean Indexing 활용
# 급여가 6000이상인 DataFrame을 필터링
# 2. 부서를 그룹화
# 3. 급여를 인덱싱
# 4. 평균 함수 적용
filtered = df3.loc[df3["Salary"] >= 6000]
filtered.groupby("Department")["Salary"].mean()
```

```python
# loc를 활용하지 않는 Boolean Indexing
# True, False를 가지는 Series를
# loc를 활용해서 할 수도 있지만

# 데이터프레임명[Boolean Indexing]
df3[df3["Salary"] >= 6000]
```

```python
# 'A' -> 65
# 'a' -> 97
# '0' -> 48
print(ord('a'))
print(ord('A'))
print(ord('0'))
# 'A' -> 'B' 
print(chr((ord('A') + 1)))

def next_char(char, num):
    return chr(ord(char) + num)

next_char('C', 3)
```

```python
str1 = "Nayeho"
result = ""
for s in str1:
    if ord(s) < ord('a'):
        result += s
    else:
        result += chr(ord(s) - (ord('a') - ord('A')))
print(result)
```

```python
titanic = pd.read_csv("titanic.csv")
titanic_df = titanic[["Survived", "Pclass", "Sex", "Age"]]
# Survived 
# 1 -> 생존, 0 -> 사망
# Pclass
# 1 -> 1등석, 2 -> 2등석, 3 -> 3등석
# print(titanic_df["Survived"].count())
# 남성 + 생존
man_survived = titanic_df[(titanic_df["Sex"] == "male") & (titanic_df["Survived"] == 1)]
print("남성 + 생존 :", man_survived["Survived"].count())
# 남성 + 사망
# 여성 + 생존
# 여성 + 사망
titanic_df.groupby(["Sex", "Survived"])["Survived"].count()
```

```python
# 1. Boolean Indexing
# 남성 + 생존
def sex_survived(sex, survived):
    return titanic_df[(titanic_df["Sex"] == sex) & (titanic_df["Survived"] == survived)]["Survived"].count()

# 남성 + 생존
print("남성 + 생존 :", sex_survived("male", 1))
print("남성 + 사망 :", sex_survived("male", 0))
print("여성 + 생존 :", sex_survived("female", 1))
print("여성 + 사망 :", sex_survived("female", 0))
```

```python
# 2.그룹화 함수 사용
# 다중 그룹화
s = titanic_df.groupby(["Sex", "Survived"])["Survived"].count()
print("남성 + 생존 :", s["male"][1])
print("남성 + 사망 :", s["male"][0])
print("여성 + 생존 :", s["female"][1])
print("여성 + 사망 :", s["female"][0])
```
