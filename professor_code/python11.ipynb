```python
# 부서별 급여 시각화
import pandas as pd
data = {
    "Department": ["HR", "IT", "IT", "Finance", "HR", "Finance"],
    "Name": ["Alice", "Bob", "Charlie", "David", "Eve", "Frank"],
    "Salary": [5000, 6000, 6500, 7000, 5200, 7100]
}
df = pd.DataFrame(data)
print(df)
```

```python
# 부서별 평균 급여 시각화
df_salary = df.groupby("Department")["Salary"].mean()
# 중복 제거된 컬럼
dept = df["Department"].unique()
plt.bar(dept, df_salary)
plt.title("Average Salary by Departments")
plt.xlabel("departmets")
plt.ylabel("Average Salary")
plt.show()
```

```python
# Pie Plot
# 부서별 인원 비율
count = df.groupby("Department")["Name"].count()
plt.pie(count, labels=count.index, autopct="%1.1f%%", startangle=90)
plt.title("Employee Distribution by Departments")
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [5, 15, 20, 25, 30]
plt.plot(x, y, marker="o")
plt.show()
```

```python
sum_salary = df.groupby("Department")["Salary"].sum()
plt.bar(sum_salary.index, sum_salary.values)
plt.show()
```

```python
lables = ["Apple", "Banana", "Cherry"]
size = [100, 150, 90]

plt.pie(size, labels=lables, autopct="%1.1f%%")
plt.show()
```

- 기초 통계량 분석

```python
df2 = pd.read_csv("sample_employee_data.csv")
df2.describe()
```

```python
# 상관관계 분석
# 상관계수 확인!!
# corr()
corr = df2[["Age", "Salary"]].corr()
print(corr)
# 1 -> 같은 컬럼
# 0.7 ~ 1 -> 강한 양의 상관관계
# 0.4 ~ 0.7 -> 보통의 양의 상관관계
# 0 ~ 0.4 -> 약한 양의 상관관계
# -0.4 ~ 0 -> 약한 음의 상관관계
# -0.7 ~ -0.4 -> 보통의 음의 상관관계
# -1 ~ -0.7 -> 강한 음의 상관관계

```
