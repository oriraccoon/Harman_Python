```python
def sub_two_number(num1, num2):
    if num1 >= num2:
        return num1 - num2
    else:
        return num2 - num1

def close10(num1, num2):
    # 내장함수 abs() : 절댓값
    return num1 if abs(num1 - 10) <= abs(num2 - 10) else num2

num1 = int(input("첫 번째 정수 입력 :"))
num2 = int(input("두 번째 정수 입력 :"))
result = close10(num1, num2)

print("10과 가까운 수 : %d"%result)
```

#### 약수를 구하는 함수!! -> 완전수(perfect number) 판별

```python
# is_divisor(num1, num2)
# num2가 num1의 약수라면 -> True 반환
# 약수가 아니라면 -> False 반환
def is_divisor(num1, num2):
    """ 
        num2가 num1의 약수인지 판단하는 함수
        약수라면 True, 아니라면 False 반환
    """
    return num1 % num2 == 0
print(is_divisor(32, 2))
print(is_divisor(32, 3))
```

```python
# 약수를 구하는 get_divisor(num1) 
# get_divisor(10)
# print -> "10의 약수 : 1 2 5 10"

def get_divisor(num1):
    print(f"{num1}의 약수 : ", end="")
    for i in range(1, num1 // 2 + 1):
        if is_divisor(num1, i):
            print(i, end=" ")
    print(num1)
```

```python
# get_sum_of_divisors(num1)
# num1의 약수의 합을 반환!!
def get_sum_of_divisors(num1):
    total = 0 # 약수의 누적 합
    for i in range(1, num1 // 2 + 1):
        if is_divisor(num1, i):
            total += i
    total += num1
    return total

print(get_sum_of_divisors(10))
print(get_sum_of_divisors(32))
```

```python
# 완전수(perfect number)
# 자기 자신을 제외한 약수의 합이
# 자기 자신과 같은 수
# 예) 10의 약수 : 1 2 5 10
# 10은 10을 제외한 약수의 합 : 8 != 10
# 10은 완전수가 아니다
# 6의 약수 : 1 2 3 6 -> 완전수
# 28의 약수 : 1 2 4 7 14 28 -> 완전수

# 완전수임을 판별하는 함수
# is_perfect_number(num1)
# num1이 완전수라면 -> True
# num2이 완전수가 아니라면 -> False
def is_perfect_number(num1):
    return get_sum_of_divisors(num1) == 2 * num1

print(is_perfect_number(6))
print(is_perfect_number(20))
print(is_perfect_number(28))
```

```python
# get_perfect_number(start, end)
# get_perfect_number(2, 1000)
# 2 ~ 1000사이의 완전수 출력!!
# 완전수 : 6 28 ??? ????

def get_perfect_number(start, end):
    print("완전수 :", end=" ")
    for i in range(start, end + 1):
        if is_perfect_number(i):
            print(i, end=" ")
    print()
get_perfect_number(2, 100000)
print("코드 끝")
```
