```python
# 최소 지폐 문제
# 38,000원이 있을 때, 만 원은 : 3장 오천 원 : 5장 천 원 : 3장
money = int(input("돈을 입력하세요 :"))
man = (money // 10000)
oh = (money % 10000) // 5000
chun = ((money % 10000) / 1000) % 5
ohbak = money % 1000 // 500
bak = money % 500 / 100
print("만원 :", man)
print("오천원 :", int(oh))
print("천원 :", int(chun))
print("오백원 :", int(ohbak))
print("백원 :", int(bak))
```

```python
# 논리연산자
# not, and, or
# 논리 not 논리 True <-> False
# 논리 and 논리
# 논리 그리고 논리
# 오늘은 2월 5일입니다 : True
# 오늘은 흐리다 
# 오늘은 2월 5일이다 그리고 오늘은 흐리다
# -> True 
# 오늘은 2월 5일이다 그리고 오늘은 흐리지 않다
# -> False 
# 오늘은 2월 6일이다 그리고 오늘은 흐리다
# -> False
# 오늘은 2월 6일이다 그리고 오늘은 흐리지 않다
# -> False
# 논리 곱
# True -> 1
# False -> 0
# 1 x 1 = 1
# 1 x 0 = 0
# 0 x 1 = 0
# 0 x 0 = 0
# True and True -> True 
# True and False -> False
# False and True -> False
# False and False -> False
# or 연산자
# 논리합
# True or True -> True
# True or False -> True
# False or True -> True
# False or False -> False
# 1 + 1 -> 2 (True)
# 1 + 0 -> 1 (True)
# 0 + 1 -> 1 (True)
# 0 + 0 -> 0 (False)

# A + 1 -> 1

print(not(5>3)) # False
print(5 > 3 and 10 > 7) # True
print(5 < 3 and 10 > 7) # False, 10>7은 연산 안해서 dead code
```

```python
# 삼항 연산자 (조건 연산자)
# a if 조건식 else b
# 조건식 : 결과가 논리!! 
# 조건식의 결과가 True(참)이면 a(if절) 실행
# 조건식의 결과가 False(거짓)이면 b(else절) 실행

score = int(input("점수를 입력하세요"))
result = "합격" if score >= 60 else "불합격"
print(result)
```

```python
f = int(input("첫 번째 정수를 입력하세요 :"))
s = int(input("두 번째 정수를 입력하세요 :"))
result = f-s if f>s else s-f
print(f"두 수의 차는 {result}입니다.")
result = f-s if f>s else -(f-s)
print(f"두 수의 차는 {result}입니다.")
```
