```python
import random

list1 = []

for i in range(10000):
    list1.append(random.randint(0, 100000))

# 오름차순 정렬(Ascending Sort)
# 1. 버블정렬(bubble sort)
# 최약체
# 시간복잡도 O^n
count = 0
while count != len(list1) - 1:
    for i in range(len(list1)-1):
        if list1[i] >= list1[i+1]:
            list1[i], list1[i+1] = list1[i+1], list1[i]
    count += 1
        
print(list1)
```

```python
list2 = []

for i in range(10000):
    list2.append(random.randint(0, 100000))
    
# 오름차순 정렬(Ascending Sort)
# 2. 선택정렬(selection sort)
# 최약체 2등
# O^n
for i in range(len(list2)-1, 0, -1):
    max_index = 0
    for j in range(i):
        if list2[j] >= list2[max_index]:
            max_index = j
    list2[max_index], list2[i] = list2[i], list2[max_index]
    
print(list2)
```

```python
list4 = [1, 2, 3, 8, 63, 4, 5, 6, 63]

for i in range(101):
    list4.append(random.randint(0, 100))

list4.sort()
# 이진탐색(Binary search)
# 정렬 필수
n = len(list4)
low_index = 0
high_index = n - 1
mid_index = (low_index + high_index) // 2
count = 0
while True:
    if 63 == list4[mid_index]:
        list4.pop(mid_index)
        count += 1
    elif 63 > list4[mid_index]:
        low_index = mid_index + 1
        mid_index = (low_index + high_index) // 2
    else:
        high_index = mid_index - 1
        mid_index = (low_index + high_index) // 2

    if low_index >= high_index:
        break

print(count)
```
