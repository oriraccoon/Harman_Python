```python
from flask import Flask, render_template, Response, request
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
import io
import base64

app = Flask(__name__)
@app.route("/")
def index():
    return render_template("index.html")

@app.route("/nogoguma")
def nogoguma():
    return render_template("nogoguma.html")

@app.route("/seoul")
def print_seoul():

    # 바탕화면에 있는 daily_weather.db를 읽어와
    conn = sqlite3.connect("dist//daily_weather.db")
    # dataframe 형태로 저장
    sql = """
    SELECT *
    FROM Seoul
    """
    df = pd.read_sql(sql, conn)
    dataframe = df.to_html()
    # 저장된 datafrmae을 웹상에 출력!!

    return render_template("seoul.html", dataframe=dataframe)
    
# request는
# 우리가 기존에 url 요청 했던 requests랑 다른 라이브러리!!

# 데이터 전송 시 
# GET, POST
# GET : url, ?key=value&key2=value2 / 공유
# POST : body / 보안

# GET방식으로 num1, num2를 받아서
# 두 수를 더해 웹 페이지에 출력
# GET, POST 방식을 통해
# 요청을 받는 경로(route)는 반드시
# 요청 방식을 선언!!
@app.route("/add", methods=["GET", "POST"])
def add_two_number():
    num1 = int(request.args["num1"])
    num2 = int(request.args["num2"])
    result = f"{num1} + {num2} = {num1 + num2}"
    return result

@app.route("/plot")
def print_plot():
    # matplotlib 활용
    # 그려진 그래프를
    # 이미지로 저장
    # 그 이미지를 웹 상에 출력
    conn = sqlite3.connect("dist//daily_weather.db")
    sql = """
    SELECT *
    FROM Seoul
    """

    df = pd.read_sql(sql, conn)
    df
    plt.figure()
    plt.title("seoul weather")
    plt.xlabel("time")
    plt.ylabel("temp")
    plt.plot(df["time"], df["temp"], marker='o')
    plt.grid()
    img = io.BytesIO()
    plt.savefig(img, format="jpg")
    Response(img.getvalue(), mimetype="image/jpg")
    img_base64 = base64.b64encode(img.getvalue()).decode('utf-8')
    plt.close()
    return f"""<!DOCTYPE html>
    <html lang="en">
    <head>
        <meta charset="UTF-8">
        <meta name="viewport" content="width=device-width, initial-scale=1.0">
        <title>Document</title>
    </head>
    <body>
        <center>
        <img src="data:image/jpeg;base64,{img_base64}" alt="Plot Image">
        <br><br>
        <button type="button" onclick="history.back()">back</button>
        </center>
    </body>
    </html>"""
    


if __name__ == "__main__":
    app.run(host="localhost")
```

```python
<!DOCTYPE html>
<html lang="ko">
<head>
    <meta charset="UTF-8">
    <title>공연 예매 달력 관리 시스템</title>
    <style>
        .container {
            display: flex;
            gap: 20px;
            padding: 20px;
            max-width: 1400px;
            margin: 0 auto;
        }

        .calendar {
            flex: 3;
            background: white;
            padding: 20px;
            border-radius: 10px;
            box-shadow: 0 0 10px rgba(0,0,0,0.1);
        }

        .calendar-table {
            width: 100%;
            border-collapse: collapse;
        }

        .calendar-table th, 
        .calendar-table td {
            padding: 15px;
            text-align: center;
            border: 1px solid #ddd;
            height: 100px;
            vertical-align: top;
            position: relative;
        }

        .date-number {
            font-weight: bold;
            margin-bottom: 5px;
        }

        .event-badge {
            background: #2196F3;
            color: white;
            padding: 3px 6px;
            border-radius: 4px;
            font-size: 0.85em;
            margin: 2px 0;
            text-align: left;
        }

        .sidebar {
            flex: 1;
            background: #f8f9fa;
            padding: 20px;
            border-radius: 10px;
            box-shadow: 0 0 10px rgba(0,0,0,0.1);
        }

        .event-item {
            margin: 10px 0;
            padding: 10px;
            background: white;
            border-radius: 5px;
            display: flex;
            justify-content: space-between;
            align-items: center;
        }

        button {
            background: #2196F3;
            color: white;
            border: none;
            padding: 8px 16px;
            border-radius: 4px;
            cursor: pointer;
        }

        input[type="text"],
        input[type="date"],
        select {
            width: 100%;
            padding: 8px;
            margin: 5px 0;
            box-sizing: border-box;
        }
    </style>
</head>
<body>
    <div class="container">
        <div class="calendar">
            <div class="calendar-header" style="position: relative;">
                <button onclick="changeMonth(-1)">◀</button>
                <h2>
                    <span id="currentYear" onclick="toggleYearList()" style="cursor: pointer;"></span>
                    <span id="currentMonth" onclick="toggleMonthList()" style="cursor: pointer;"></span>
                </h2>
                <button onclick="changeMonth(1)">▶</button>

                <div id="yearList" style="display: none; position: absolute; background: white; border: 1px solid #ddd; border-radius: 5px; max-height: 200px; overflow-y: auto; z-index: 10;"></div>
                <div id="monthList" style="display: none; position: absolute; background: white; border: 1px solid #ddd; border-radius: 5px; max-height: 200px; overflow-y: auto; z-index: 10;"></div>
            </div>

            <table class="calendar-table" id="calendarTable">
                <thead>
                    <tr>
                        <th>일</th><th>월</th><th>화</th><th>수</th><th>목</th><th>금</th><th>토</th>
                    </tr>
                </thead>
                <tbody id="calendarBody"></tbody>
            </table>
        </div>

        <div class="sidebar">
            <h3>공연 목록 (최대 20개)</h3>
            <div class="event-list" id="eventList"></div>

            <div style="margin-top: 20px;">
                <h4>새 공연 추가</h4>
                <select id="genreSelect" onchange="updateShowList()">
                    <option value="">장르 선택</option>
                </select>
                <select id="showSelect">
                    <option value="">공연 선택</option>
                </select>
                <input type="text" id="eventType" placeholder="장르 직접 입력">
                <input type="text" id="eventName" placeholder="공연명 직접 입력">
                <input type="date" id="eventDate">
                <button onclick="addEvent()">추가</button>
            </div>

            <div style="margin-top: 20px;">
                <button onclick="saveCheckState()">체크 상태 저장</button>
                <button onclick="loadCheckState()">저장된 목록 불러오기</button>
            </div>
        </div>
    </div>

    <script>
        let events = [];
        let genres = {};
        let currentDate = new Date();
        const MAX_EVENTS = 20;

        function generateCalendar() {
            const calendarBody = document.getElementById('calendarBody');
            calendarBody.innerHTML = '';
            const year = currentDate.getFullYear();
            const month = currentDate.getMonth();
            document.getElementById('currentYear').textContent = `${year}년`;
            document.getElementById('currentMonth').textContent = `${month + 1}월`;

            const firstDay = new Date(year, month, 1).getDay();
            const lastDate = new Date(year, month + 1, 0).getDate();
            let date = 1;

            for (let i = 0; i < 6; i++) {
                let row = '<tr>';
                for (let j = 0; j < 7; j++) {
                    if (i === 0 && j < firstDay) {
                        row += '<td></td>';
                    } else if (date > lastDate) {
                        break;
                    } else {
                        row += `<td data-date="${year}-${String(month + 1).padStart(2, '0')}-${String(date).padStart(2, '0')}">
                            <div class="date-number">${date}</div>
                        </td>`;
                        date++;
                    }
                }
                row += '</tr>';
                calendarBody.innerHTML += row;
                if (date > lastDate) break;
            }
            renderEvents();
        }

        function renderEvents() {
            document.querySelectorAll('.calendar-day').forEach(cell => {
                cell.innerHTML = `<div class="date-number">${cell.getAttribute('data-date').split('-')[2]}</div>`;
            });

            events.forEach(event => {
                const cell = document.querySelector(`.calendar-day[data-date="${event.date}"]`);
                if (cell) {
                    const badge = document.createElement('div');
                    badge.className = 'event-badge';
                    badge.textContent = `${event.type} - ${event.name}`;
                    cell.appendChild(badge);
                }
            });
        }

        function addEvent() {
            const genre = document.getElementById('genreSelect').value || document.getElementById('eventType').value;
            const name = document.getElementById('showSelect').value || document.getElementById('eventName').value;
            const date = document.getElementById('eventDate').value;

            if (!genre || !name || !date) {
                alert('모든 필드를 입력하세요');
                return;
            }

            if (events.length >= MAX_EVENTS) {
                alert('최대 20개까지만 추가할 수 있습니다');
                return;
            }

            events.push({ id: Date.now(), type: genre, name: name, date: date, checked: true });
            updateEventList();
            generateCalendar();
        }

        function updateEventList() {
            const eventList = document.getElementById('eventList');
            eventList.innerHTML = events.map(event => `
                <div class="event-item">
                    <label>
                        <input type="checkbox" ${event.checked ? 'checked' : ''} onchange="toggleEvent(${event.id})">
                        ${event.date} <${event.type}> ${event.name}
                    </label>
                    <button onclick="deleteEvent(${event.id})">삭제</button>
                </div>
            `).join('');
        }

        function toggleEvent(id) {
            const event = events.find(e => e.id === id);
            event.checked = !event.checked;
            generateCalendar();
        }

        function deleteEvent(id) {
            events = events.filter(e => e.id !== id);
            updateEventList();
            generateCalendar();
        }

        function updateShowList() {
            const genre = document.getElementById('genreSelect').value;
            const showSelect = document.getElementById('showSelect');
            showSelect.innerHTML = '<option value="">공연 선택</option>';
            if (genres[genre]) {
                genres[genre].forEach(show => {
                    const option = document.createElement('option');
                    option.value = show;
                    option.textContent = show;
                    showSelect.appendChild(option);
                });
            }
        }

        generateCalendar();
    </script>
</body>
</html>
```
