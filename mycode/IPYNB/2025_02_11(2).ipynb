```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]

plt.figure()
plt.plot(x, y, 'r', marker='v')
plt.title("simple line plot", fontsize = 15, color = 'blue')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.show()
```

```python
import pandas as pd

data = {
    "Department": ["HR", "IT", "IT", "Finance", "HR", "Finance"],
    "Name": ["Alice", "Bob", "Charlie", "David", "Eve", "Frank"],
    "Salary": [5000, 6000, 6500, 7000, 5200, 7100]
}

df = pd.DataFrame(data)

df_dms = df.groupby("Department")["Salary"].mean()
dpt = df["Department"].unique()



plt.figure()
plt.bar(dpt, df_dms)
plt.title("DMS")
plt.xlabel("Department")
plt.ylabel("Salary")
plt.grid()
plt.show()
```

```python
numb = df.groupby("Department")["Name"].count()

plt.pie(numb, labels=numb.index, autopct="%2.1f%%", startangle=90)
plt.title("number of Employee each Department")
plt.show()
```

```python
x = [1,2,3,4,5]
y = [5,15,20,25,30]

plt.figure()
plt.plot(x,y, marker='o')
plt.show()
```

```python
sum_salary = df.groupby("Department")["Salary"].sum()
plt.bar(sum_salary.index, sum_salary.values)
plt.show()
```

```python
lables = ["Apple", "Banana", "Cherry"]
size = [100, 150, 90]

plt.pie(size, labels=lables, autopct="%1.1f%%", startangle = 90)
plt.show()
```
