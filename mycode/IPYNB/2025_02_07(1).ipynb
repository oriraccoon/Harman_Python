```python
def cal(num1, num2, op):
    if op == "-":
        return num1 - num2
    else:
        return num1 + num2
```

```python
def close_to_ten(num1, num2):
    if num1 >= 10: dis1 = num1 - 10
    else: dis1 = 10 - num1

    if num2 >= 10: dis2 = num2 - 10
    else: dis2 = 10 - num2

    if dis1 <= dis2: return num1
    else: return num2
```

```python
def close_to_ten(num1,num2):
    if abs(num1 - 10) <= abs(num2 - 10): return num1
    else: return num2
```

```python
num1 = int(input("first num :"))
num2 = int(input("second num :"))
result = close_to_ten(num1, num2)

print("10에 가까운 수 : %d"%result)
```

```python
def get_divisor1(num1):
    div = []
    for i in range(1, num1 + 1 // 2):
        if num1 % i == 0: div.append(i)
    div.append(num1)
    print(div)

get_divisor1(100)
```

```python
def get_divisor2(num1):
    for i in range(1, num1 + 1 // 2):
        if num1 % i == 0: print(i, end=" ")
    print(num1)

get_divisor2(100)
```

```python
# 완전수(perfect number)
# 자기 자신을 제외한 약수의 합이
# 자기 자신과 같은 수
# 예) 10의 약수 : 1 2 5 10
# 10은 10을 제외한 약수의 합 : 8 != 10
# 10은 완전수가 아니다
# 6의 약수 : 1 2 3 6 -> 완전수
# 28의 약수 : 1 2 4 7 14 28 -> 완전수

def is_perfect_number(num1):
    divSum = 0
    for i in range(1, num1):
        if num1 % i == 0: divSum += i
    if num1 == divSum:
        return True
    else:
        return False

print(is_perfect_number(10))
```
