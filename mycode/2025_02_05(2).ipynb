```python
score = int(input("점수입력 :"))
# 90점 이상이면 -> "A"
# 80 ~ 89점 -> "B"
# 70 ~ 79점 -> "C"
# 60 ~ 69점 -> "D"
# ~ 59점 -> "F"
if score >= 90:
    print("A")
elif score >= 80:
    print("B")
elif score >= 70:
    print("C")
elif score >= 60:
    print("D")
else:
    print("F")
```

```python
money = int(input("금액을 입력하세요 :")) # 8500
print("[1]아침햇살(1000원) [2]솔의눈(700원) [3]데자와(500원)")
menu = int(input("메뉴를 고르세요 :")) # 1
# 사용자가 1을 입력하면 -> 아침햇살!!!
while menu == 1 or menu == 2 or menu == 3:
    if menu == 1:
        print("아침햇살!!")
        money -= 1000
        print(f"잔돈 : {money}")
    elif menu == 2:
        print("솔의눈!!")
        money -= 700
        print(f"잔돈 : {money}")
    elif menu == 3:
        print("데자와!!")
        money -= 500
        print(f"잔돈 : {money}")
    else:
        print("취소")

    print("[1]아침햇살(1000원) [2]솔의눈(700원) [3]데자와(500원)")
    menu = int(input("메뉴를 고르세요 :")) # 1        
    # 2를 입력하면 -> 솔의눈!!!
    # 3을 입력하면 -> 데자와!!!
    # 사용자가 입력한 금액에서
    # 해당 메뉴의 가격을 빼고 출력
    # 잔돈 : 7500원
    # 오천원 : 1장
    # 천원 : 2장
    # 오백원 : 1개
    # 백원 : 0개
print(f"잔돈 : {money}")
print(f"오천원 : {money // 5000}장\n천원 : {money % 5000 // 1000}장\n오백원 : {money % 1000 // 500}개\n백원 : {money % 500 // 100}개")
```
